# Introduction

In this example, the problem of the system of two spheres of radious $R$ in a box of size $L$ is approached using Metropolis-Hasting method. This is an problem that is not easy to solve by other MC techniques but Metropolis method can be used to statisticaly study some variables of the system (e.g. distance between the spheres or the minimum distance to a wall).

# Problem Description

We are going to consider the simplest case of two hard spheres of radius $R$ within a squared box of side $L$ centred at $(0,0)$. Each state can be described by the coordinates of the centers of the spheres, which will be referred as $\bf{x}_1$ and $\bf{x}_2$. Hence, the probability density function (*pdf*) can be formulated as function of the centres coordinates as $\rho(\bf{x}_1, \bf{x}_2)$. Even though $\rho(\bf{x}_1, \bf{x}_2)$ will not be straightforward to obtain analitically for  all the variable space, given the position centres ${\bf{x}}^i_1$ and ${\bf{x}}^i_2$ for a certain state $i$, the probability density $\rho({\bf{x}}^i_1, {\bf{x}}^i_2)$ can be computed:

* $\rho({\bf{x}}^i_1, {\bf{x}}^i_2)=0$ if spheres overlap (i.e. $d({\bf{x}}^i_1, {\bf{x}}^i_2) < 2R$) or if any of them are partially outside the box (i.e. distance to a wall is less than $R$).
* $\rho({\bf{x}}^i_1, {\bf{x}}^i_2)=\mathrm{constant}$ otherwise, for all the allowed states, the specific value of the constant depend on normalization but it will not be required for the application of this method.

which will allow a trivial calculation of the acceptance ratio of a certain state according to Metropolis-Hasting method. It will be one if
it is an allowed state and null if not allowed.

# Method Implementation

A program to solve the problem described in the previous section by using Metropolis-Hastings method has been developed in Python and it is provided together with this notebook (i.e. *metropolis.py*). In this section, an overview of how Metropolis-Hasting algorithm has been applied to this problem is given.

Initially, several systems (i.e. walkers) have to be initialized an arbitrary point of the phase space, by selecting two arbitrary initial coordinates for the sphere centres. In the program, this points can be selected by the user or it will be automatically sampled from a uniform distribution until initial state is valid. Because of the characteristics of the system, a uniform 2D distribution will share some properties with the objective distribution $\rho(\bf{x}_1, \bf{x}_2)$, so convergence will be much faster when its used as initial point.

The, for each of the walkers, the system variables have to be randomly perturbed. In this case, the coordinates of the center of one of the sphere will be displace according to a uniform distribution centred in zero and with a maximum jump size provided by the user (or a default of the order of $R$).

In order to illustrate both the problem and the method, an animation has been created that represents how the position of the spheres in the box change randomly with time for each walker.

```python
from metropolis import two_spheres_metropolis
from anim_example import animation_metropolis
import numpy as np
np.random.seed(seed=10) # to obtain reproducible results

L = 1.0 # box dimensions
R = 0.10 # shere radious
n_walkers = 1 # number of walkers to simulate
w_centres = [-0.2,0.0, 0.2, 0.0 ] # initial sphere position
step_max = 0.10 # max possible move in 1D in one step

metropolis = two_spheres_metropolis(L, R, n_walkers, w_centres)

from IPython.display import HTML
HTML(animation_metropolis(metropolis, frames = 200,
                          fps = 10, step_max = step_max))
```

As it can be seen in the animation, the spheres start at an arbitrary position and they move randomly around the box. When the Metropolis-Hastings method is used, several walkers (i.e. systems of two spheres in a box in this case) will be initialized and randomly perturbed.  After enough time, the state of each walker will not depend on the initialization conditions and the ensemble of walkers will be consistent with
the objective *pdf*, the state of each walker will be a random state following  $\rho(\bf{x}_1, \bf{x}_2)$.

# Results

At first, the convergence of the ensemble of walkers from different fixed initial position to the objective probablility distribution function $\rho(\bf{x}_1, \bf{x}_2)$ will be tested and compared when valid initial states are sampled from a uniform distribution.

This will be carried out by controlling how two variables (mean distance between centers and mean minimum distance to wal) that depend on the *pdf* evolve with the number of Metropolis-Hastings steps from the initial position. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from metropolis import two_spheres_metropolis
np.random.seed(seed=10) # to obtain reproducible results
 
L = 1.0 # box dimensions
R = 0.10 # shere radious
n_walkers = 10000 # number of walkers

# n_iterations = n_steps*step_width
step_width = 5
n_steps = 1000

# initial sphere positions
w_centres = [[-0.2,0.0, 0.2, 0.0 ],
             [-0.35,0.35, 0.35, 0.35 ],
             [-0.35,-0.15, -0.15, -0.35 ],
             None] # will be assigned randomly if None
n_ensembles = len(w_centres)

# arrays to save results
step_arr = np.linspace(0,n_steps*step_width, n_steps)
dinter_arr = np.zeros((n_steps, n_ensembles))
dwall_arr = np.zeros((n_steps, n_ensembles))

# init and perform Metropolis-Hastings steps for all the ensembles
for ens in xrange(n_ensembles):
    metropolis = two_spheres_metropolis(L, R, n_walkers,  w_centres = w_centres[ens])
    for i in xrange(n_steps):
        dinter_arr[i, ens] = np.mean(metropolis.inter_distance())
        dwall_arr[i, ens] = np.mean(metropolis.wall_distance())
        metropolis.step(n_steps=step_width)

# plot results
lbl = ["Fixed init", "Uniform init" ]
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(1,2,1)
ax1.set_title("Mean distance between centres")
lines1_fixed = ax1.plot(step_arr, dinter_arr[:,0:3], label =  lbl[0])
lines1_random = ax1.plot(step_arr, dinter_arr[:,3], label =  lbl[1])
leg1 = ax1.legend()
ax2 = fig.add_subplot(1,2,2)
ax2.set_title("Mean minimum distance to wall")
lines2_fixed = ax2.plot(step_arr, dwall_arr[:,0:3], label =  lbl[0])
lines2_random = ax2.plot(step_arr, dwall_arr[:,3], label =  lbl[1])
leg2 = ax2.legend()
```

Previous figures show how the statistical variables of the system studied converge to the same value independently of how each walker was initialized after enough steps. Because a large number of walkers was used (i.e. 10000), the statistical fluctuations from the mean once its converged are very small. For the slowest ensemble, it took around 1500 steps to converge. As expected, when walkers are initialized in valid states sampled from an uniform distribution convergence is much faster because the resulting distribution is very close to $\rho(\bf{x}_1, \bf{x}_2)$.

At the previous study, the maximum that a sphere could move in each direction (i.e. x and y) was $R/\pi$, the characteristics of the distribution used to perturb the system affect convergence speed. In the fowolling example, several values will be used for the maximum displacement $\Delta$ in the random walk.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from metropolis import two_spheres_metropolis
np.random.seed(seed=10) # to obtain reproducible results
 
L = 1.0 # box dimensions
R = 0.10 # shere radious
n_walkers = 10000 # number of walkers

# n_iterations = n_steps*step_width
step_width = 5
n_steps = 1000

# initial sphere positions
w_centres = [-0.2,0.0, 0.2, 0.0 ] # will be assigned randomly if None
step_max = [0.1*R, 0.5*R, R, 5*R, 10*R]
n_ensembles = len(step_max)

# arrays to save results
step_arr = np.linspace(0,n_steps*step_width, n_steps)
dinter_arr = np.zeros((n_steps, n_ensembles))
dwall_arr = np.zeros((n_steps, n_ensembles))

# init and perform Metropolis-Hastings steps for all the ensembles
for ens in xrange(n_ensembles):
    metropolis = two_spheres_metropolis(L, R, n_walkers,  w_centres = w_centres)
    for i in xrange(n_steps):
        dinter_arr[i, ens] = np.mean(metropolis.inter_distance())
        dwall_arr[i, ens] = np.mean(metropolis.wall_distance())
        metropolis.step(n_steps=step_width, step_max=step_max[ens])

# plot results
lbl = ["$\Delta$ = {:.2f}*R".format(step_max[ens]/R) for ens in xrange(n_ensembles)]
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(1,2,1)
ax1.set_title("Mean distance between centres")
lines1 = [ax1.plot(step_arr, dinter_arr[:,ens], label =  lbl[ens]) for ens in xrange(n_ensembles)]
leg1 = ax1.legend(loc = 4)
ax1.set_xscale("log")
ax2 = fig.add_subplot(1,2,2)
ax2.set_title("Mean minimum distance to wall")
lines2 = [ax2.plot(step_arr, dinter_arr[:,ens], label =  lbl[ens]) for ens in xrange(n_ensembles)]
leg2 = ax2.legend(loc = 4)
ax2.set_xscale("log")
```

As expected, different step size distributions for the random walk cause different convergence speeds. If $\Delta$ is too small, the convergence is very slow (blue line does not even converge in the range used). If $\Delta$ is large (e.g. $\Delta=10$, purple line) many moves will be rejected slowing convergence. The optimal value could be determined this way but it will depend on $R$ and $L$ (and the number of particles if more than two are consided). Aditionally, other random step function could be considered as a Gaussian. For the rest of the study, a $\Delta$ of approximaterly the value $R$ will be used. 

Apart from the mean of the variables, any quantitity which depends on $\rho(\bf{x}_1, \bf{x}_2)$ could be estimated. For example, the walkers could be sampled to estimate the histogram of the distance between the spheres or the the minimum distance 
to wall:

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from metropolis import two_spheres_metropolis
np.random.seed(seed=10) # to obtain reproducible results
 
L = 1.0 # box dimensions
R = [0.001, 0.01, 0.05, 0.10, 0.20 ] # sphere radious
n_walkers = 10000 # number of walkers

n_iterations = 1000

n_ensembles = len(R)

# to save results
inter_distance =  np.zeros((n_walkers, n_ensembles))
wall_distance =  np.zeros((n_walkers, n_ensembles))


# init and perform Metropolis-Hastings steps for all the ensembles
for ens in xrange(n_ensembles):
    metropolis = two_spheres_metropolis(L, R[ens], n_walkers)
    metropolis.step(n_steps=n_iterations)
    inter_distance[:,ens] = metropolis.inter_distance()
    wall_distance[:,ens] = metropolis.wall_distance()
    
# plot results
lbl = ["R = {:.3f}".format(R[ens]) for ens in xrange(n_ensembles)]
nbins = 50
fig = plt.figure(figsize=(20,8))
ax1 = fig.add_subplot(1,2,1)
values1, bins1, patches1 = ax1.hist(inter_distance, bins=nbins, histtype='step', label= lbl)
leg1 = ax1.legend()
ax2 = fig.add_subplot(1,2,2)
values2, bins2, patches2 = ax2.hist(wall_distance, bins=nbins,histtype='step', label= lbl)
leg2 = ax2.legend()
```

In the previous figures, histograms of the distance between spheres and minimum distance to wall distance distributions are provided for systems where the sphere radious $R$ was 0.2, 0.1, 0.05, 0.01 and 0.001 (box size was $L=1.0$). A total of 10000 walkers was used for each case.

In adddition, the Metropolis-Hasting method will allow to estimate the probability distribution function $\rho(\bf{x}_1, \bf{x}_2)$ by means of a 4D histogram (which will not be easy to plot here) among other quantities. For example, the next example provides a 2D histogram of the sphere positions (probability of finding a sphere centre in a region of the box) for diferent radious.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from metropolis import two_spheres_metropolis
np.random.seed(seed=10) # to obtain reproducible results
 
L = 1.0 # box dimensions
R = [0.05, 0.10, 0.15] # sphere radious
n_walkers = 100000 # number of walkers

n_iterations = 1000
n_ensembles = len(R)

# prepare plots
lbl = ["R = {:.3f}".format(R[ens]) for ens in xrange(n_ensembles)]
fig, axarr = plt.subplots(1,3, figsize=(20,6))

# init and perform Metropolis-Hastings steps for all the ensembles
for ens in xrange(n_ensembles):
    metropolis = two_spheres_metropolis(L, R[ens], n_walkers)
    metropolis.step(n_steps=n_iterations)
    axarr[ens].set_title(lbl[ens])
    x_values = np.abs(np.concatenate((metropolis.w_c[:,0],metropolis.w_c[:,2])))
    y_values = np.abs(np.concatenate((metropolis.w_c[:,1],metropolis.w_c[:,3])))
    (counts, xedges, yedges, image) = axarr[ens].hist2d(x_values, y_values, range = [[0, 0.5], [0, 0.5]])
    image.set_cmap("Greys")

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
fig.colorbar(image, cax=cbar_ax)
```

Because of the simmetries of the system, all the centre positions have been reflected to the upper right quarter so larger statistics are avaliable. A total of 100000 walkers were generated for $R$ of 0.5, 0.10 and 0.15. As expected, for small spheres $R<0.05$ the probability of finding a sphere at a certial is aproximately uniform. However, when larger spheres are used it is less likely to find a sphere at the center, so $\rho(\bf{x}_1, \bf{x}_2)$ differs more from uniform distribution.
