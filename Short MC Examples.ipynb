# Approximation of $\pi$ using MC methods

```python
import numpy as np
np.random.seed(seed=27) # to obtain reproducible results

N = 1000000 # number of samples
x = 2*np.random.rand(N) - 1
y = 2*np.random.rand(N) - 1
# check inside circle and estimate pi
pi_aprox = 4.*np.sum((x**2+y**2) < 1)/N

print "Aproximation for pi using {:.2e} samples: {:.5f} ".format(N,pi_aprox)
```

# Calculation of integrals using MC methods

$$\int_0^1 x {\rm{d}}x = \frac{1}{2}$$
$$\int_0^1 x^2 {\rm{d}}x = \frac{1}{3}$$
$$\int_0^1 \sin(\pi x) {\rm{d}}x = \frac{2}{\pi}$$

```python
import numpy as np
np.random.seed(seed=27) # to obtain reproducible results

N = 1000000 # number of samples
x = np.random.rand(N) # sample x values uniformly
# apply functions
f_1 = x
f_2 = x**2
f_3 = np.sin(x*np.pi)
# sum results ( calculate integrals )
int_1 = np.sum(f_1)/N
int_2 = np.sum(f_2)/N
int_3 = np.sum(f_3)/N
# estimate errors
err_1 = np.std(f_1)/np.sqrt(N)
err_2 = np.std(f_2)/np.sqrt(N)
err_3 = np.std(f_3)/np.sqrt(N)


print "integral - f(x) = x - [0,1]         :" \
      "{:.5f} \pm {:.5f}".format(int_1, err_1)
print "integral - f(x) = x**2 - [0,1]      :" \
      "{:.5f} \pm {:.5f}".format(int_2, err_2)
print "integral - f(x) = sin(x*pi) - [0,1] :" \
      "{:.5f} \pm {:.5f}".format(int_3, err_3)
```

# Random number by composite method

$$\rho(x) = a \exp(-\lambda_1 x)+ b \exp(-\lambda_2 x)$$

$$\rho(x) = f_1(x)+f_2(x)$$

$$F(x) = F_1(x)+ F_2(x)$$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(seed=27) # to obtain reproducible results

a = 10.
b = 1.
l_1 = 5.
l_2 = 10.


N = 1000000 # number of samples
u = np.random.rand(N) # generate uniform random number
x = np.random.rand(N) # generate uniform random number
F_1 = - (1./l_1)*np.log(1-x)
F_2 = - (1./l_2)*np.log(1-x)
treshold = a/(a+b)
condition = u < treshold
# F_1 when condition T, F_2 when false
F = np.where(condition, F_1, F_2) 

# plot and compare with pdf
n, bins, patches = plt.hist(F, bins=100,histtype='step', normed=True)
x_pdf = np.linspace(np.min(bins),np.max(bins), 10000)
norm = a/l_1 + b/l_2
rho_pdf = (a*np.exp(-l_1*x_pdf)+b*np.exp(-l_2*x_pdf))/norm
plt.plot(x_pdf,rho_pdf)
plt.yscale("log")
plt.ylim((1e-5,20))
plt.ylabel("Normalized distribution")
plt.xlabel("x value")
```

# Normal Random Number (Box-Muller)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(seed=27) # to obtain reproducible results

N = 1000000 # number of samples
u_1 = np.random.rand(N) # generate uniform random number
u_2 = np.random.rand(N) # generate uniform random number
# change of coordinates
r_squared = -2.*np.log(u_1)
theta = 2*np.pi*u_2
# z_0 and z_1 are independent normal random vars
z_0 = np.sqrt(r_squared)*np.cos(theta)
z_1 = np.sqrt(r_squared)*np.sin(theta)

# plot and compare with pdf
n, bins, patches = plt.hist(z_0, bins=100, histtype='step', normed=True)
n, bins, patches = plt.hist(z_1, bins=100, histtype='step', normed=True)
x_pdf = np.linspace(np.min(bins),np.max(bins), 10000)
pdf = np.exp(-x_pdf**2/2)/np.sqrt(2*np.pi)
plt.plot(x_pdf, pdf)
plt.ylabel("Normalized distribution")
plt.xlabel("x value")

# check independence
print "Correlation coef: {:4.2e}".format(np.corrcoef(z_0,z_1)[0,1])
```

# Random Numbers by Rejection Sampling

$$\rho(x)=1-|x|$$



```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(seed=27) # to obtain reproducible results

N = 1000000 # number of samples
u_1 = np.random.rand(N) # generate uniform random number
u_2 = np.random.rand(N) # generate uniform random number
x = 2.*u_1 - 1. # uniform between -1 and 1
g_x = 1. # have to be larger than f(x) for all x
f_x = 1. - np.abs(x)
# get accepted values
samples = x[u_2<(f_x/g_x)] 

n, bins, patches = plt.hist(samples, bins=100, histtype='step',normed=True)
x_pdf = np.linspace(np.min(bins),np.max(bins), 10000)
pdf = 1 - np.abs(x_pdf)
plt.plot(x_pdf, pdf, "-r")
plt.ylabel("Normalized distribution")
plt.xlabel("x value")
```
